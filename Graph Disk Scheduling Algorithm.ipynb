```python
from matplotlib import pyplot as plt
import math
```

```python
request_list = [106, 10, 31, 186, 78, 52, 113, 167, 104, 198]
curr_head = 50
max_cylinder = 199
```

```python
def SCAN(request_list,curr_head,moving_direction):
    temp_list = request_list.copy()
    left_request = []
    right_request = []
    x = []
    y = []
    x.append(curr_head)
    if(moving_direction=="Left"):
        for i in temp_list:
            if i<curr_head:
                left_request.append(i)
            else:
                right_request.append(i)
        left_request.sort(reverse = True)
        for i in left_request:
            x.append(i)
        x.append(0)
        right_request.sort()
        for i in right_request:
            x.append(i)
    elif(moving_direction=="Right"):
        for i in temp_list:
            if i>curr_head:
                right_request.append(i)
            else:
                left_request.append(i)
        right_request.sort()
        for i in right_request:
            x.append(i)
        x.append(max_cylinder)
        left_request.sort(reverse = True)
        for i in left_request:
            x.append(i)
    
    size = len(x)
    for i in range(0,size):
        y.append(-i)

    plt.figure(figsize=(10, 5))
    plt.xticks(range(0, max_cylinder + 1, 20))
    plt.plot(x,y, color="green", markerfacecolor = 'blue', marker='o', markersize = 5, linewidth = 2, label="SCAN")
    plt.ylim = (0,size)
    plt.xlim = (0,max_cylinder)
    plt.yticks([])
    plt.title("SCAN Disk Scheduling Algorithm")
    plt.show()
```

```python
SCAN(request_list,curr_head,"Left")
```
